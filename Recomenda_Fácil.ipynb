```python
import matplotlib.pyplot as plt
import numpy as np

# Define a função fornecida pelo usuário
def gerar_recomendacoes(historico):
    n = len(historico)
    return 2 * n - 1

# Define um intervalo de valores para n (número de compras)
# Vamos considerar de 0 a 15 compras para visualização
n_values = np.arange(0, 16)

# Para usar a função, simulei um histórico de compras de tamanho n
recomendacoes_values = []
for n in n_values:
  # Criei uma lista fictícia de tamanho n para simular o histórico
  historico_ficticio = ['item'] * n
  recomendacoes = gerar_recomendacoes(historico_ficticio)
  recomendacoes_values.append(recomendacoes)


# Cria o gráfico
plt.figure(figsize=(10, 6))
plt.plot(n_values, recomendacoes_values, marker='o', linestyle='-')
plt.title('Número de Recomendações vs. Número de Compras Anteriores')
plt.xlabel('Número de Compras Anteriores (n)')
plt.ylabel('Número de Recomendações Geradas')
plt.grid(True)
plt.xticks(n_values)
plt.show()
```
